```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot
sns.scatterplot(data=df, x="G", y="W")
plt.show()

# Histograms
sns.histplot(df["G"], kde=True)
plt.show()

sns.histplot(df["W"], kde=True)
plt.show()

# Histogram grouped by category
sns.histplot(data=df, x="G", hue="CONF", multiple="stack")
plt.show()
```
